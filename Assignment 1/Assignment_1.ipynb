# Machine Learning with K-Nearest Neighbors (KNN)
## Part 1: KNN with Iris Dataset
This section loads the classic Iris dataset and applies K-Nearest Neighbors classification.


```python

# Import necessary libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = datasets.load_iris()
data, labels = iris.data, iris.target

# Split the dataset into training and testing sets (80-20 split)
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, train_size=0.8, test_size=0.2, random_state=12)

# Initialize and train the K-Nearest Neighbors classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(train_data, train_labels)

# Predict on test data
iris_predictions = knn.predict(test_data)

# Compute accuracy
iris_accuracy = accuracy_score(iris_predictions, test_labels)

print(f"Iris Dataset KNN Accuracy: {iris_accuracy:.4f}")
```

## Part 2: KNN with Simulated Dataset
This section generates a synthetic dataset with three classes and applies KNN classification.

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Generate a different simulated dataset
data, labels = make_classification(
    n_samples=150,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    n_classes=3,
    random_state=1
)

# Step 2: Split the data into training and testing sets
train_data, test_data, train_labels, test_labels = train_test_split(
    data, labels, train_size=0.8, test_size=0.2, random_state=12
)

# Step 3: Evaluate KNN for different values of K
k_values = range(1, 21)
train_accuracies = []
test_accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(train_data, train_labels)

    train_predictions = knn.predict(train_data)
    train_accuracy = accuracy_score(train_predictions, train_labels)
    train_accuracies.append(train_accuracy)

    test_predictions = knn.predict(test_data)
    test_accuracy = accuracy_score(test_predictions, test_labels)
    test_accuracies.append(test_accuracy)

    # Print accuracies for each K
    print(f"K={k}: Train Accuracy={train_accuracy:.4f}, Test Accuracy={test_accuracy:.4f}")

# Step 4: Plotting the accuracies
plt.figure(figsize=(10, 6))
plt.plot(k_values, train_accuracies, marker='o', label='Training Accuracy')
plt.plot(k_values, test_accuracies, marker='x', label='Testing Accuracy')
plt.title('KNN Accuracy for Different Values of K')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# Step 5: Visualization of final KNN with optimal K
optimal_k = k_values[test_accuracies.index(max(test_accuracies))]
knn_optimal = KNeighborsClassifier(n_neighbors=optimal_k)
knn_optimal.fit(train_data, train_labels)
final_test_predictions = knn_optimal.predict(test_data)

plt.figure(figsize=(8, 6))
plt.scatter(test_data[:, 0], test_data[:, 1], c=final_test_predictions, cmap='viridis', marker='o', edgecolor='k', label='Predicted')
plt.scatter(train_data[:, 0], train_data[:, 1], c=train_labels, cmap='viridis', marker='x', label='Training Data')
plt.title(f"KNN Classification with Optimal K={optimal_k}")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
